```python
import pandas as pd, numpy as np, joblib, matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
```

```python
LABEL_COL  = None  # set to actual label if you have it later, e.g. "IsFraud"
CONTAM     = 0.08  # expected fraud rate (tune)
df = pd.read_csv("/home/mubasshir/Desktop/Research/Insurance/DataSet_Patient/Enhanced_Health/enhanced_health_insurance_claims.csv")
```

```python
# ====== LIGHT FEATURE ENGINEERING (claim-level only) ======
```

```python
if "ClaimDate" in df.columns:
    df["ClaimDate"] = pd.to_datetime(df["ClaimDate"], errors="coerce", infer_datetime_format=True)
    df["claim_year"] = df["ClaimDate"].dt.year
    df["claim_month"] = df["ClaimDate"].dt.month
    df["claim_dayofweek"] = df["ClaimDate"].dt.dayofweek
else:
    df["claim_year"] = df["claim_month"] = df["claim_dayofweek"] = np.nan

# Numeric casts
for col in ["ClaimAmount", "PatientAge", "PatientIncome"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Log amount
if "ClaimAmount" in df.columns:
    df["log_amount"] = np.log1p(df["ClaimAmount"].clip(lower=0))
else:
    df["log_amount"] = 0.0
```

```python
# Rarity of codes (inverse frequency)
for col in ["DiagnosisCode", "ProcedureCode"]:
    if col in df.columns:
        freq = df[col].value_counts(dropna=True)
        df[f"{col}_rarity"] = df[col].map(lambda x: -np.log(freq.get(x, 1) / len(df)) if pd.notna(x) else 0.0).fillna(0.0)
    else:
        df[f"{col}_rarity"] = 0.0
```

```python
# ====== FEATURE MATRIX ======
num_feats = [c for c in ["ClaimAmount","log_amount","PatientAge","PatientIncome",
                         "claim_year","claim_month","claim_dayofweek",
                         "DiagnosisCode_rarity","ProcedureCode_rarity"]
             if c in df.columns]

cat_feats = [c for c in ["DiagnosisCode","ProcedureCode","PatientGender",
                         "ClaimStatus","PatientMaritalStatus","PatientEmploymentStatus",
                         "ClaimType","ClaimSubmissionMethod"]
             if c in df.columns]

X = df[num_feats + cat_feats].copy()

pre = ColumnTransformer([
    ("num", SimpleImputer(strategy="median"), num_feats),
    ("cat", Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("enc", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
    ]), cat_feats)
])
```

```python
# ====== MODEL (Isolation Forest) ======
iso = Pipeline([
    ("pre", pre),
    ("model", IsolationForest(n_estimators=400, contamination=CONTAM, random_state=42))
])
iso.fit(X)

Z = iso["pre"].transform(X)
scores = -iso["model"].score_samples(Z)          # higher => more anomalous
decision = iso["model"].decision_function(Z)     # < 0 => outlier
flags = (decision < 0).astype(int)
```

```python
# ====== SAVE RESULTS ======
out_claims = df.copy()
out_claims["anomaly_score"] = scores
out_claims["anomaly_flag"]  = flags
out_csv = f"/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/result_datasets/Patient_base_output/claims_with_anomaly_scores.csv"
out_claims.to_csv(out_csv, index=False)
print(f"Saved claim scores -> {out_csv}")
```

```python
# Save model
model_path = f"/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/result_datasets/Patient_base_output/model.pkl"
joblib.dump(iso, model_path)
print(f"Saved model -> {model_path}")
```

```python
# ====== OPTIONAL EVALUATION (only if you add a label later) ======
if LABEL_COL is not None and LABEL_COL in df.columns:
    y = df[LABEL_COL]
    # if labels are strings like Yes/No/True/False, map to 1/0
    if y.dtype == "O":
        y = y.astype(str).str.strip().str.lower().map({"yes":1,"true":1,"1":1,"fraud":1,"no":0,"false":0,"0":0}).fillna(0).astype(int)
    # accuracy using anomaly_flag as prediction (unsupervised => crude)
    acc = accuracy_score(y, flags)
    print(f"Accuracy (unsupervised flags vs label): {acc:.4f}")
    # ROC using anomaly scores
    try:
        auc = roc_auc_score(y, scores)
        fpr, tpr, _ = roc_curve(y, scores)
        plt.figure()
        plt.plot(fpr, tpr, label=f"AUC={auc:.2f}")
        plt.plot([0,1],[0,1],"--")
        plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
        plt.title("ROC (Isolation Forest scores)")
        plt.legend()
        roc_path = f"{OUT_DIR}/roc_curve.png"
        plt.savefig(roc_path, bbox_inches="tight")
        print(f"Saved ROC curve -> {roc_path}")
    except Exception as e:
        print("ROC not computed:", e)
else:
    print("No label column provided -> skipped accuracy/ROC (unsupervised scoring only).")
```

```python
import os, json, numpy as np, pandas as pd, joblib, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import (
    accuracy_score, roc_auc_score, roc_curve,
    precision_recall_curve, average_precision_score, confusion_matrix
)
```

```python
# ===== CONFIG =====
CSV_PATH    = "/home/mubasshir/Desktop/Research/Insurance/DataSet_Patient/Enhanced_Health/enhanced_health_insurance_claims.csv"   # your data (same schema as training)
MODEL_PATH  = "/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/result_datasets/Patient_base_output/model.pkl"   # saved pipeline from training
OUT_DIR     = "/home/mubasshir/Desktop/Research/Insurance/health-insurance-fraud-detection/result_datasets/Patient_base_output/fraud_artifacts_claim_only"             # where to write outputs
LABEL_COL   = None   # put your label later, e.g. "IsFraud"; keep None if unlabeled
TOP_K       = 100    # when unlabeled, save top-K most suspicious claims
```

```python
# ===== LOAD =====
os.makedirs(OUT_DIR, exist_ok=True)
pipe = joblib.load(MODEL_PATH)      # this is a sklearn Pipeline with ("pre", ...), ("model", IsolationForest)
df = pd.read_csv(CSV_PATH).copy()
```

```python
# ===== LIGHT FEATURES (must match training) =====
```

```python
# Dates
if "ClaimDate" in df.columns:
    df["ClaimDate"] = pd.to_datetime(df["ClaimDate"], errors="coerce", infer_datetime_format=True)
    df["claim_year"] = df["ClaimDate"].dt.year
    df["claim_month"] = df["ClaimDate"].dt.month
    df["claim_dayofweek"] = df["ClaimDate"].dt.dayofweek
else:
    df["claim_year"] = df["claim_month"] = df["claim_dayofweek"] = np.nan

```

```python
# Numeric
for col in ["ClaimAmount","PatientAge","PatientIncome"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
df["log_amount"] = np.log1p(df.get("ClaimAmount", 0)).astype(float)

```

```python
# Rarity (inverse frequency) for codes
for col in ["DiagnosisCode","ProcedureCode"]:
    if col in df.columns:
        freq = df[col].value_counts(dropna=True)
        df[f"{col}_rarity"] = df[col].map(lambda x: -np.log(freq.get(x, 1)/len(df)) if pd.notna(x) else 0.0).fillna(0.0)
    else:
        df[f"{col}_rarity"] = 0.0
```

```python
num_feats = [c for c in [
    "ClaimAmount","log_amount","PatientAge","PatientIncome",
    "claim_year","claim_month","claim_dayofweek",
    "DiagnosisCode_rarity","ProcedureCode_rarity"
] if c in df.columns]
cat_feats = [c for c in [
    "DiagnosisCode","ProcedureCode","PatientGender",
    "ClaimStatus","PatientMaritalStatus","PatientEmploymentStatus",
    "ClaimType","ClaimSubmissionMethod"
] if c in df.columns]
X = df[num_feats + cat_feats].copy()
```

```python
# ===== SCORE WITH SAVED MODEL =====
Z = pipe["pre"].transform(X)
scores   = -pipe["model"].score_samples(Z)          # higher => more suspicious
decision =  pipe["model"].decision_function(Z)      # < 0 => outlier
flags    = (decision < 0).astype(int)

scored = df.copy()
scored["anomaly_score"] = scores
scored["anomaly_flag"]  = flags
scored_path = Path(OUT_DIR) / "scored_claims.csv"
scored.to_csv(scored_path, index=False)
print(f"Saved scores -> {scored_path}")
```

```python
SAVE_PLOTS = True 
plt.figure()
pd.Series(scores).dropna().hist(bins=50)
plt.title("Anomaly Score Distribution (higher = more suspicious)")
plt.xlabel("anomaly_score"); plt.ylabel("count")
if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "hist_anomaly_score.png"), bbox_inches="tight")
plt.show()

# 2) Flag distribution
plt.figure()
pd.Series(flags).value_counts().sort_index().plot(kind="bar")
plt.title("Anomaly Flags (0=normal, 1=flagged)")
plt.xlabel("anomaly_flag"); plt.ylabel("count")
if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "bar_flags.png"), bbox_inches="tight")
plt.show()

# 3) ClaimAmount vs score (if ClaimAmount present)
if "ClaimAmount" in df.columns:
    plt.figure()
    plt.scatter(df["ClaimAmount"], scores, s=8, alpha=0.5)
    plt.title("ClaimAmount vs Anomaly Score")
    plt.xlabel("ClaimAmount"); plt.ylabel("anomaly_score")
    if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "scatter_amount_vs_score.png"), bbox_inches="tight")
    plt.show()

# 4) Top-10 DiagnosisCode by mean score
if "DiagnosisCode" in df.columns:
    diag_means = scored.groupby("DiagnosisCode")["anomaly_score"].mean().sort_values(ascending=False).head(10)
    plt.figure()
    diag_means.plot(kind="bar")
    plt.title("Top-10 DiagnosisCode by Mean Anomaly Score")
    plt.xlabel("DiagnosisCode"); plt.ylabel("mean anomaly_score"); plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "bar_diagcode_mean_score.png"), bbox_inches="tight")
    plt.show()

# 5) Top-10 ProcedureCode by mean score
if "ProcedureCode" in df.columns:
    proc_means = scored.groupby("ProcedureCode")["anomaly_score"].mean().sort_values(ascending=False).head(10)
    plt.figure()
    proc_means.plot(kind="bar")
    plt.title("Top-10 ProcedureCode by Mean Anomaly Score")
    plt.xlabel("ProcedureCode"); plt.ylabel("mean anomaly_score"); plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "bar_proccode_mean_score.png"), bbox_inches="tight")
    plt.show()

# 6) Monthly trend (if ClaimDate present)
if "ClaimDate" in df.columns:
    dt = pd.to_datetime(df["ClaimDate"], errors="coerce")
    monthly = scored.assign(_m=dt.dt.to_period("M").astype(str)).groupby("_m")["anomaly_score"].mean().dropna()
    if not monthly.empty:
        plt.figure()
        monthly.plot(kind="line", marker="o")
        plt.title("Monthly Mean Anomaly Score")
        plt.xlabel("Month"); plt.ylabel("mean anomaly_score"); plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "line_monthly_mean_score.png"), bbox_inches="tight")
        plt.show()

# ===== OPTIONAL: if you later add labels, show ROC/PR without saving JSON =====
if LABEL_COL is not None and LABEL_COL in df.columns:
    y = df[LABEL_COL]
    if y.dtype == "O":
        y = y.astype(str).str.strip().str.lower().map(
            {"yes":1,"true":1,"1":1,"fraud":1,"no":0,"false":0,"0":0}
        ).fillna(0).astype(int)

    auc = roc_auc_score(y, scores)
    fpr, tpr, _ = roc_curve(y, scores)
    pr, rc, _ = precision_recall_curve(y, scores)
    ap = average_precision_score(y, scores)
    print(f"ROC-AUC: {auc:.4f} | PR-AUC: {ap:.4f}")

    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC={auc:.3f}")
    plt.plot([0,1],[0,1],"--")
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC (Isolation Forest scores)"); plt.legend()
    if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "roc_curve.png"), bbox_inches="tight")
    plt.show()

    plt.figure()
    plt.plot(rc, pr)
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"PR Curve (AP={ap:.3f})")
    if SAVE_PLOTS: plt.savefig(os.path.join(OUT_DIR, "pr_curve.png"), bbox_inches="tight")
    plt.show()
```
