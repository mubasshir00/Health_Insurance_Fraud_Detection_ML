```python
# ===========================
# Leakage-Free Accuracy Boost (refined)
# ===========================
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, average_precision_score, precision_recall_curve, roc_curve
from sklearn.calibration import CalibratedClassifierCV
from scipy.stats import randint, uniform

# 1) Load
DATA_PATH = "/home/mubasshir/Desktop/Research/Insurance/DataSet_Patient/Medicare_Physician_Data/CMSData_sampled.csv"  # change if needed
df = pd.read_csv(DATA_PATH, dtype={"Rndrng_Prvdr_Zip5": str}, low_memory=False).fillna(0)

# 2) Drop identifiers / text
drop_cols = [
    "Rndrng_Prvdr_Last_Org_Name","Rndrng_Prvdr_First_Name","Rndrng_Prvdr_MI",
    "Rndrng_Prvdr_Crdntls","Rndrng_Prvdr_St1","Rndrng_Prvdr_St2",
    "Rndrng_Prvdr_City","Rndrng_Prvdr_Cntry","Rndrng_Prvdr_RUCA_Desc",
    "Rndrng_Prvdr_Zip5","Rndrng_NPI"
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# 3) Synthetic label (same as before)  EXCLUDE defining cols from X
df["Fraud_Label"] = np.where(
    (df["Avg_Sbmtd_Chrg"] > 5 * df["Avg_Mdcr_Alowd_Amt"]) |
    (df["Tot_Srvcs"] > df["Tot_Srvcs"].quantile(0.99)), 1, 0
).astype(int)

# 4) SAFE feature engineering that does NOT peek at y
#    a) simple signals (OK on full df, not derived from y)
if "Avg_Mdcr_Pymt_Amt" in df.columns and "Avg_Mdcr_Stdzd_Amt" in df.columns:
    df["Payment_Stdzd_Diff"] = df["Avg_Mdcr_Pymt_Amt"] - df["Avg_Mdcr_Stdzd_Amt"]
df["Inpatient_Flag"] = np.where(df.get("Place_Of_Srvc","O")=="F", 1, 0)
if "Tot_Bene_Day_Srvcs" in df.columns and "Tot_Benes" in df.columns:
    df["Bene_Day_Density"] = df["Tot_Bene_Day_Srvcs"]/(df["Tot_Benes"]+1)

#    b) DO NOT create HCPCS aggregates/freq on full data (leakage)  ### CHANGE
#       We'll do these AFTER the train/test split from TRAIN only.

#    c) Categorical columns
cat_cols = [c for c in [
    "Rndrng_Prvdr_Gndr","Rndrng_Prvdr_Ent_Cd","Rndrng_Prvdr_State_Abrvtn",
    "Rndrng_Prvdr_Type","HCPCS_Cd","HCPCS_Drug_Ind","Place_Of_Srvc"
] if c in df.columns]

# 5) Build X (drop leakage sources)
leak_cols = [c for c in [
    "Avg_Sbmtd_Chrg","Avg_Mdcr_Alowd_Amt","Tot_Srvcs",
    "Charge_Allowed_Ratio","Charge_Payment_Diff","Services_Per_Beneficiary",
    "Service_Day_Ratio","High_Cost_Flag"
] if c in df.columns]
y = df["Fraud_Label"].copy()
X = df.drop(columns=["Fraud_Label"] + leak_cols)

# 6) Train/valid split
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 7) Frequency encode categoricals using TRAIN distribution only
X_train = X_train_raw.copy()
X_test  = X_test_raw.copy()

for col in cat_cols:
    if col in X_train.columns:
        # TRAIN-only normalized freq
        freq = X_train[col].astype(str).value_counts(normalize=True)
        X_train[col] = X_train[col].astype(str).map(freq).fillna(0)
        X_test[col]  = X_test[col].astype(str).map(freq).fillna(0)

# 7b) HCPCS aggregates using TRAIN only (no leakage)  ### CHANGE
if "HCPCS_Cd" in X_train.columns and "Avg_Mdcr_Pymt_Amt" in X_train.columns:
    train_grp = X_train_raw.assign(Avg_Mdcr_Pymt_Amt=X_train_raw.get("Avg_Mdcr_Pymt_Amt", 0))
    # Use the ORIGINAL train rows for robust stats (before logs/clips)
    gmean = X_train_raw.groupby("HCPCS_Cd")["Avg_Mdcr_Pymt_Amt"].mean()
    gstd  = X_train_raw.groupby("HCPCS_Cd")["Avg_Mdcr_Pymt_Amt"].std().fillna(0)
    gfreq = X_train_raw["HCPCS_Cd"].astype(str).value_counts()

    for df_part in (X_train, X_test):
        df_part["HCPCS_pay_mean"] = df_part["HCPCS_Cd"].map(gmean).fillna(0)
        df_part["HCPCS_pay_std"]  = df_part["HCPCS_Cd"].map(gstd).fillna(0)
        df_part["HCPCS_Cd_freq"]  = df_part["HCPCS_Cd"].astype(str).map(gfreq).fillna(0)

# 8) Numeric cleanup: drop non-numeric, winsorize, add log1p
nn_train = X_train.select_dtypes(include=["object"]).columns.tolist()
nn_test  = X_test.select_dtypes(include=["object"]).columns.tolist()
if nn_train: X_train = X_train.drop(columns=nn_train)
if nn_test:  X_test  = X_test.drop(columns=nn_test)

# winsorize to tame outliers
low_clip = X_train.quantile(0.01); high_clip = X_train.quantile(0.99)
X_train = X_train.clip(lower=low_clip, upper=high_clip, axis=1)
X_test  = X_test.clip(lower=low_clip, upper=high_clip, axis=1)

# add log1p on skewed positives
for c in ["Avg_Mdcr_Pymt_Amt","Avg_Mdcr_Stdzd_Amt","Tot_Benes","Tot_Bene_Day_Srvcs",
          "Payment_Stdzd_Diff","HCPCS_pay_mean","HCPCS_pay_std","HCPCS_Cd_freq"]:
    if c in X_train.columns:
        X_train[c+"_log1p"] = np.log1p(np.maximum(X_train[c], 0))
        X_test[c+"_log1p"]  = np.log1p(np.maximum(X_test[c], 0))

# 9) NO SCALING for HGB (trees dont need it)  ### CHANGE
X_train_s = X_train.to_numpy()
X_test_s  = X_test.to_numpy()

# 10) Strong learner + randomized hyperparameter search
hgb = HistGradientBoostingClassifier(
    early_stopping=True, validation_fraction=0.1, random_state=42
)

param_dist = {
    "learning_rate":  uniform(0.03, 0.15),     # 0.030.18
    "max_depth":      randint(3, 10),
    "max_iter":       randint(200, 700),
    "l2_regularization": uniform(0.0, 0.5),
    "min_samples_leaf":  randint(10, 200)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
search = RandomizedSearchCV(
    estimator=hgb,
    param_distributions=param_dist,
    n_iter=25,
    scoring="roc_auc",
    cv=cv,
    verbose=1,
    n_jobs=-1,
    random_state=42
)
search.fit(X_train_s, y_train)

best_clf = search.best_estimator_
print("Best params:", search.best_params_)
print("CV best AUC:", search.best_score_)

# 10b) Probability calibration (train-only)  ### CHANGE
calib = CalibratedClassifierCV(best_clf, method="isotonic", cv=3)
calib.fit(X_train_s, y_train)

# 11) Predict probabilities on test
y_prob = calib.predict_proba(X_test_s)[:, 1]  # calibrated probs

# 12) Threshold search that maximizes accuracy (ties by recall)
thresholds = np.linspace(0.0, 1.0, 2001)
best = {"thr":0.5,"acc":-1,"prec":0,"rec":0}
from sklearn.metrics import precision_score, recall_score
for t in thresholds:
    y_pred = (y_prob >= t).astype(int)
    acc = accuracy_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred, zero_division=0)
    if (acc > best["acc"]) or (acc == best["acc"] and rec > best["rec"]):
        best = {"thr":t, "acc":acc,
                "prec":precision_score(y_test, y_pred, zero_division=0),
                "rec":rec}

print(f"\nChosen threshold: {best['thr']:.4f}")
print(f"Accuracy: {best['acc']:.4%} | Precision: {best['prec']:.4%} | Recall: {best['rec']:.4%}")

# 13) Final metrics and plots
y_pred_final = (y_prob >= best["thr"]).astype(int)
print("\nClassification Report at chosen threshold:\n",
      classification_report(y_test, y_pred_final, digits=4))

print("ROC-AUC (calibrated):", roc_auc_score(y_test, y_prob))
print("PR-AUC (calibrated):",  average_precision_score(y_test, y_prob))

fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(); plt.plot(fpr, tpr, label="HGB tuned + calibrated"); plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC (tuned)")
plt.legend(); plt.show()

prec, rec, _ = precision_recall_curve(y_test, y_prob)
plt.figure(); plt.plot(rec, prec, label="HGB tuned + calibrated")
plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("PR Curve (tuned)")
plt.legend(); plt.show()

# 14) Permutation importance (unchanged)
from sklearn.inspection import permutation_importance
idx = np.random.RandomState(42).choice(len(X_test_s), size=min(8000, len(X_test_s)), replace=False)
pi = permutation_importance(calib, X_test_s[idx], y_test.iloc[idx],
                            scoring="roc_auc", n_repeats=5, random_state=42)
imp = pd.Series(pi.importances_mean, index=X_train.columns).sort_values(ascending=False)
print("\nTop features (permutation AUC importance):\n", imp.head(15))
```

```python
# ===========================
# Recall-Boosted Pipeline (edits on your code)
# ===========================
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import (accuracy_score, classification_report, roc_auc_score,
                             average_precision_score, precision_recall_curve, roc_curve,
                             precision_score, recall_score, f1_score, confusion_matrix)
from sklearn.calibration import CalibratedClassifierCV
from scipy.stats import randint, uniform

# ======= Recall target & weighting (Q1) =======  ### CHANGE
TARGET_RECALL = 0.70          # set your Q1 target here
POS_WEIGHT_EXP = 1.0          # 1.0 = classic inverse frequency; <1 softens, >1 strengthens

# 1) Load
DATA_PATH = "/home/mubasshir/Desktop/Research/Insurance/DataSet_Patient/Medicare_Physician_Data/CMSData_sampled.csv"
df = pd.read_csv(DATA_PATH, dtype={"Rndrng_Prvdr_Zip5": str}, low_memory=False).fillna(0)

# 2) Drop identifiers / text
drop_cols = [
    "Rndrng_Prvdr_Last_Org_Name","Rndrng_Prvdr_First_Name","Rndrng_Prvdr_MI",
    "Rndrng_Prvdr_Crdntls","Rndrng_Prvdr_St1","Rndrng_Prvdr_St2",
    "Rndrng_Prvdr_City","Rndrng_Prvdr_Cntry","Rndrng_Prvdr_RUCA_Desc",
    "Rndrng_Prvdr_Zip5","Rndrng_NPI"
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# 3) Synthetic label (same as before)
df["Fraud_Label"] = np.where(
    (df["Avg_Sbmtd_Chrg"] > 5 * df["Avg_Mdcr_Alowd_Amt"]) |
    (df["Tot_Srvcs"] > df["Tot_Srvcs"].quantile(0.99)), 1, 0
).astype(int)

# 4) SAFE feature engineering (unchanged)
if "Avg_Mdcr_Pymt_Amt" in df.columns and "Avg_Mdcr_Stdzd_Amt" in df.columns:
    df["Payment_Stdzd_Diff"] = df["Avg_Mdcr_Pymt_Amt"] - df["Avg_Mdcr_Stdzd_Amt"]
df["Inpatient_Flag"] = np.where(df.get("Place_Of_Srvc","O")=="F", 1, 0)
if "Tot_Bene_Day_Srvcs" in df.columns and "Tot_Benes" in df.columns:
    df["Bene_Day_Density"] = df["Tot_Bene_Day_Srvcs"]/(df["Tot_Benes"]+1)

cat_cols = [c for c in [
    "Rndrng_Prvdr_Gndr","Rndrng_Prvdr_Ent_Cd","Rndrng_Prvdr_State_Abrvtn",
    "Rndrng_Prvdr_Type","HCPCS_Cd","HCPCS_Drug_Ind","Place_Of_Srvc"
] if c in df.columns]

# 5) Build X (drop leakage sources)
leak_cols = [c for c in [
    "Avg_Sbmtd_Chrg","Avg_Mdcr_Alowd_Amt","Tot_Srvcs",
    "Charge_Allowed_Ratio","Charge_Payment_Diff","Services_Per_Beneficiary",
    "Service_Day_Ratio","High_Cost_Flag"
] if c in df.columns]
y = df["Fraud_Label"].copy()
X = df.drop(columns=["Fraud_Label"] + leak_cols)

# 6) Split
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 7) Train-only encodings (no leakage)
X_train = X_train_raw.copy()
X_test  = X_test_raw.copy()

for col in cat_cols:
    if col in X_train.columns:
        freq = X_train[col].astype(str).value_counts(normalize=True)
        X_train[col] = X_train[col].astype(str).map(freq).fillna(0)
        X_test[col]  = X_test[col].astype(str).map(freq).fillna(0)

# 7b) HCPCS aggregates from TRAIN only
if "HCPCS_Cd" in X_train.columns and "Avg_Mdcr_Pymt_Amt" in X_train.columns:
    gmean = X_train_raw.groupby("HCPCS_Cd")["Avg_Mdcr_Pymt_Amt"].mean()
    gstd  = X_train_raw.groupby("HCPCS_Cd")["Avg_Mdcr_Pymt_Amt"].std().fillna(0)
    gfreq = X_train_raw["HCPCS_Cd"].astype(str).value_counts()
    for df_part in (X_train, X_test):
        df_part["HCPCS_pay_mean"] = df_part["HCPCS_Cd"].map(gmean).fillna(0)
        df_part["HCPCS_pay_std"]  = df_part["HCPCS_Cd"].map(gstd).fillna(0)
        df_part["HCPCS_Cd_freq"]  = df_part["HCPCS_Cd"].astype(str).map(gfreq).fillna(0)

# 8) Numeric cleanup, winsorize, log1p
nn_train = X_train.select_dtypes(include=["object"]).columns.tolist()
nn_test  = X_test.select_dtypes(include=["object"]).columns.tolist()
if nn_train: X_train = X_train.drop(columns=nn_train)
if nn_test:  X_test  = X_test.drop(columns=nn_test)

low_clip = X_train.quantile(0.01); high_clip = X_train.quantile(0.99)
X_train = X_train.clip(lower=low_clip, upper=high_clip, axis=1)
X_test  = X_test.clip(lower=low_clip, upper=high_clip, axis=1)

for c in ["Avg_Mdcr_Pymt_Amt","Avg_Mdcr_Stdzd_Amt","Tot_Benes","Tot_Bene_Day_Srvcs",
          "Payment_Stdzd_Diff","HCPCS_pay_mean","HCPCS_pay_std","HCPCS_Cd_freq"]:
    if c in X_train.columns:
        X_train[c+"_log1p"] = np.log1p(np.maximum(X_train[c], 0))
        X_test[c+"_log1p"]  = np.log1p(np.maximum(X_test[c], 0))

X_train_s = X_train.to_numpy()
X_test_s  = X_test.to_numpy()

# ======= Class weighting for recall (sample_weight) =======  ### CHANGE
n_pos = (y_train == 1).sum()
n_neg = (y_train == 0).sum()
pos_weight = (n_neg / max(1, n_pos)) ** POS_WEIGHT_EXP    # e.g., if 1:10 imbalance  ~10
w_train = np.where(y_train.values == 1, pos_weight, 1.0).astype(float)

# 10) Strong learner + randomized hyperparameter search
hgb = HistGradientBoostingClassifier(
    early_stopping=True, validation_fraction=0.1, random_state=42
)

param_dist = {
    "learning_rate":  uniform(0.03, 0.15),
    "max_depth":      randint(3, 10),
    "max_iter":       randint(200, 700),
    "l2_regularization": uniform(0.0, 0.5),
    "min_samples_leaf":  randint(10, 200)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Use PR-AUC to prioritize positives  ### CHANGE
search = RandomizedSearchCV(
    estimator=hgb,
    param_distributions=param_dist,
    n_iter=25,
    scoring="average_precision",   # prefer recall/precision on positives
    cv=cv,
    verbose=1,
    n_jobs=-1,
    random_state=42
)
# Pass sample_weight into CV fits  ### CHANGE
search.fit(X_train_s, y_train, **{"sample_weight": w_train})

best_clf = search.best_estimator_
print("Best params:", search.best_params_)
print("CV best PR-AUC:", search.best_score_)   # changed metric name  ### CHANGE

# 10b) Probability calibration with weights  ### CHANGE
calib = CalibratedClassifierCV(best_clf, method="isotonic", cv=3)
calib.fit(X_train_s, y_train, sample_weight=w_train)

# 11) Predict probabilities on test
y_prob = calib.predict_proba(X_test_s)[:, 1]

# ======= Threshold selection to hit TARGET_RECALL =======  ### CHANGE
def sweep_thresholds(y_true, p, steps=2000):
    thr = np.linspace(0.0, 1.0, steps+1)
    rows = []
    for t in thr:
        yhat = (p >= t).astype(int)
        rows.append({
            "thr": t,
            "recall": recall_score(y_true, yhat, zero_division=0),
            "precision": precision_score(y_true, yhat, zero_division=0),
            "accuracy": accuracy_score(y_true, yhat),
            "f1": f1_score(y_true, yhat, zero_division=0),
            "alerts": int(yhat.sum())
        })
    return pd.DataFrame(rows)

grid = sweep_thresholds(y_test, y_prob)

# pick smallest alerting threshold that meets recall, break ties by higher precision then accuracy
meet = grid[grid["recall"] >= TARGET_RECALL]
if not meet.empty:
    chosen = meet.sort_values(["alerts","precision","accuracy"], ascending=[True, False, False]).iloc[0]
else:
    # fallback: best recall achievable, tie-break by precision then accuracy
    chosen = grid.sort_values(["recall","precision","accuracy"], ascending=[False, False, False]).iloc[0]

tstar = chosen["thr"]
print(f"\n[Recall-oriented] Chosen threshold: {tstar:.4f}  "
      f"| Recall: {chosen['recall']:.4%}  | Precision: {chosen['precision']:.4%}  "
      f"| Accuracy: {chosen['accuracy']:.4%}  | Alerts: {int(chosen['alerts'])}")

# 13) Final metrics & plots at recall-optimized threshold
y_pred_final = (y_prob >= tstar).astype(int)
print("\nClassification Report at chosen threshold:\n",
      classification_report(y_test, y_pred_final, digits=4))

print("ROC-AUC (calibrated):", roc_auc_score(y_test, y_prob))
print("PR-AUC  (calibrated):", average_precision_score(y_test, y_prob))

# Small preview of operating curve near the chosen point
print("\nTop operating points near chosen threshold:")
print(grid.iloc[(grid['thr']-tstar).abs().argsort()[:8]][
    ["thr","recall","precision","accuracy","f1","alerts"]
].to_string(index=False))

# Curves (unchanged)
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(); plt.plot(fpr, tpr, label="HGB tuned + calibrated"); plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate"); plt.title("ROC")
plt.legend(); plt.show()

prec, rec, _ = precision_recall_curve(y_test, y_prob)
plt.figure(); plt.plot(rec, prec, label="HGB tuned + calibrated")
plt.axvline(TARGET_RECALL, linestyle="--", alpha=0.6, label=f"Target recall={TARGET_RECALL:.2f}")
plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("PR Curve")
plt.legend(); plt.show()

# 14) Permutation importance (works with calibrated model)
from sklearn.inspection import permutation_importance
idx = np.random.RandomState(42).choice(len(X_test_s), size=min(8000, len(X_test_s)), replace=False)
pi = permutation_importance(calib, X_test_s[idx], y_test.iloc[idx],
                            scoring="average_precision", n_repeats=5, random_state=42)  # ### CHANGE
imp = pd.Series(pi.importances_mean, index=X_train.columns).sort_values(ascending=False)
print("\nTop features (permutation PR-AUC importance):\n", imp.head(15))
```
